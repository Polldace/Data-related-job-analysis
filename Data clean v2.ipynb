```python
import pandas as pd
import numpy as np
```

* Eliminate duplication

```python
dfdata = pd.read_excel('cleandata.xlsx',index_col = 0)
dfdata.shape
```

```python
# Dropping any duplicate rows:
scrape_data = dfdata.drop(columns='job_url')
scrape_data = scrape_data.drop_duplicates()
scrape_data.shape
```

```python
# Reset index
scrape_data.reset_index(drop = True, inplace=True)
scrape_data.head(50)
```

* Format salary column

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(scrape_data["job_salary"].value_counts())
```

```python
#some targeted cleaning of salary information to make parsing easier 
#—> remove “\n”, “$”, and “,”
scrape_data["job_salary"] = scrape_data["job_salary"].str.replace('\n', '')
scrape_data["job_salary"] = scrape_data["job_salary"].str.replace(',', '')
scrape_data["job_salary"] = scrape_data["job_salary"].str.replace('$', '')
```

```python
# Add a new column 'unit_salary_period' after 'job_salary'
scrape_data.insert(loc = scrape_data.columns.get_loc("job_salary")+1
, column="unit_salary_period", value=np.nan)
```

```python
# Fill 'unit_salary_period' with salary unit: per hour/day/annum
scrape_data.loc[(scrape_data["job_salary"].notnull())&(scrape_data["job_salary"].str.contains('p'))&(scrape_data["job_salary"].str.contains('a')), "unit_salary_period"] = 'per annum'
scrape_data.loc[(scrape_data["job_salary"].notnull())&(scrape_data["job_salary"].str.contains('p'))&(scrape_data["job_salary"].str.contains('h')), "unit_salary_period"] = 'per hour'
scrape_data.loc[(scrape_data["job_salary"].notnull())&(scrape_data["job_salary"].str.contains('p'))&(scrape_data["job_salary"].str.contains('d')), "unit_salary_period"] = 'per day'


scrape_data.head(50)
```

```python
# Adjust columns order
# cols = scrape_data.columns.tolist()
# cols
# cols =['job_title',
#  'job_company',
#  'job_location',
#  'job_salary',
#  'unit_salary_period',
#  'job_date',
#  'job_indus',
#  'job_info']

# scrape_data = scrape_data[cols]
# scrape_data.head(30)
```

```python
# Split salary range according to '-'
scrape_data["job_salary"] = scrape_data["job_salary"].str.replace('/','-')

scrape_data['low_salary'] =scrape_data["job_salary"].str.split('-').str[0]
scrape_data['high_salary'] =scrape_data["job_salary"].str.split('-').str[1]

# Extract only numerical values in low and high salary columns
scrape_data['low_salary'] = scrape_data['low_salary'].str.extract('(\d+)', expand=False).astype(float)
scrape_data['high_salary'] = scrape_data['high_salary'].str.extract('(\d+)', expand=False).astype(float)
scrape_data.head(20)
```

```python
scrape_data['revised_salary'] = np.nan
```

```python
# Calculate mean according to the range of salary
for i in range(len(scrape_data)):
    if ~np.isnan(scrape_data.loc[i,'low_salary']) and ~np.isnan(scrape_data.loc[i,'high_salary']):
        scrape_data.loc[i,'revised_salary'] = (scrape_data.loc[i, 'low_salary'] + scrape_data.loc[i, 'high_salary'])/2
    else:
        scrape_data.loc[i,'revised_salary'] = scrape_data.loc[i, 'low_salary']
    
```

```python
scrape_data.loc[(scrape_data['unit_salary_period'].notnull())&(scrape_data['unit_salary_period'].str.contains('per day')),'revised_salary'] = scrape_data.loc[(scrape_data['unit_salary_period'].notnull())&(scrape_data['unit_salary_period'].str.contains('per day')),'revised_salary'] *260
scrape_data.loc[(scrape_data['unit_salary_period'].notnull())&(scrape_data['unit_salary_period'].str.contains('per hour')),'revised_salary']= scrape_data.loc[(scrape_data['unit_salary_period'].notnull())&(scrape_data['unit_salary_period'].str.contains('per hour')),'revised_salary']*2080
```

```python
scrape_data.loc[(scrape_data['revised_salary'].notnull())&(scrape_data['revised_salary']<500),'revised_salary'] = scrape_data.loc[(scrape_data['revised_salary'].notnull())&((scrape_data['revised_salary'])<500),'revised_salary']*1000
```

```python
scrape_data.to_excel('cleandatav2.xlsx')
```

* Categorise job title

```python
scrape_data["job_title"] = scrape_data["job_title"].str.lower()
with pd.option_context('display.max_rows', None, 'display.max_columns', 3):
    print(scrape_data["job_title"].value_counts())
```

```python
scrape_data.insert(loc = scrape_data.columns.get_loc("job_title")+1, 
                   column = "jobtitle_group", value=np.nan)
scrape_data.head(20)
```

```python
# Categorise job title into several types

scrape_data.loc[scrape_data["job_title"].str.contains('business analyst|commercial analyst|ba|business analytics'),"jobtitle_group"]='Business Analyst'
scrape_data.loc[(scrape_data["job_title"].str.contains('data') & scrape_data["job_title"].str.contains('analyst|analytics')),"jobtitle_group"]='Data Analyst'
scrape_data.loc[scrape_data["job_title"].str.contains('data engineer'),"jobtitle_group"]='Data Engineer'
scrape_data.loc[scrape_data["job_title"].str.contains('data scientist'),"jobtitle_group"]='Data Scientist'

scrape_data.loc[(scrape_data["job_title"].str.contains('intelligence|bi')&scrape_data["jobtitle_group"].isnull()),"jobtitle_group"]='Business Intelligence'
scrape_data.loc[scrape_data["job_title"].str.contains('consultant'),"jobtitle_group"]='Consultant'

scrape_data.loc[(scrape_data["job_title"].str.contains('analyst')&scrape_data["jobtitle_group"].isnull()),"jobtitle_group"]='Other Analyst'
# scrape_data.loc[scrape_data["job_title"].str.contains('BI')&scrape_data["jobtitle_group"].isnull(),"jobtitle_group"]='BI Developer'
#scrape_data.loc[scrape_data["jobtitle_group"].isnull(),"jobtitle_group"]=scrape_data.loc[scrape_data["jobtitle_group"].isnull(),"job_title"]

scrape_data["jobtitle_group"].value_counts()
```

```python
# Remove na of jobtitle_group from dataframe, and reset index
scrape_data = scrape_data.dropna(subset = ['jobtitle_group'])
scrape_data.reset_index(drop = True, inplace=True)
scrape_data.head(50)
```

```python
scrape_data.to_excel('cleandatav3.xlsx')
```
